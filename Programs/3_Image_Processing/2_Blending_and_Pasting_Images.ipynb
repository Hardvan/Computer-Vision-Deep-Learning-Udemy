## `Blending` Images of the `Same Size`

### Resizing Images

```python
img1 = cv2.resize(img1, (1200, 1200))
img2 = cv2.resize(img2, (1200, 1200))
```

### Blending Images
Formula: `new_pixel` = `alpha` * `pixel1` + `beta` * `pixel2` + `gamma`

```python
blended = cv2.addWeighted(src1=img1,
                          alpha=0.8, # Weight of the first image
                          src2=img2,
                          beta=0.1, # Weight of the second image
                          gamma=10 # Scalar added to each sum
                          )

plt.imshow(blended)
```

## Overlay a small image on top of a larger image (No Blending)

```python
img1 = cv2.imread("./images/00-puppy.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("./images/watermark_no_copy.png")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600, 600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0 # Offset from the left
y_offset = 0 # Offset from the top
```

```python
x_end = x_offset + small_img.shape[1] # Width of the small image
y_end = y_offset + small_img.shape[0] # Height of the small image
```

```python
small_img.shape
```

### Overlaying images

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img

plt.imshow(large_img)
```
