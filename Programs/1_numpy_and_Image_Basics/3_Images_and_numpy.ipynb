## Converting Image to a numpy array

- Both np.array and np.asarray will create a numpy array from an image.
- Differenc: **np.asarray()** will **not create a copy** of the image if the image is already a numpy array.  
This is important to know because if you are working with a large image, you may not want to create a copy of the image in memory.

```python
img_arr = np.asarray(img)
img_arr
```

```python
type(img_arr)
```

```python
img_arr.shape
```

**imshow** is a function from **matplotlib** that will **display an image**.

```python
plt.imshow(img_arr)
```

## RGB

### Red Channel

```python
plt.imshow(img_arr[:, :, 0])
```

with cmap

```python
plt.imshow(img_arr[:, :, 0], cmap="gray")
```

### Green Channel

```python
plt.imshow(img_arr[:, :, 1], cmap="gray")
```

### Blue Channel

```python
plt.imshow(img_arr[:, :, 2], cmap="gray")
```

### Setting Green Channel to 0

```python
img_arr_copy = img_arr.copy()
img_arr_copy[:, :, 1] = 0
plt.imshow(img_arr_copy)
```
