# `Detecting Fraud Notes using Keras`

```python
import numpy as np
from numpy import genfromtxt
```

```python
data = genfromtxt("../DATA/bank_note_data.txt", delimiter=",")
```

```python
data
```

## Separating labels and features

```python
labels = data[:, 4]
labels
```

```python
features = data[:, :4]
features
```

```python
X = features
y = labels
```

## Splitting data into train and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.33,
                                                    random_state=42)
```

```python
len(X_train)
```

## Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
scaled_X_train.max()
```

```python
scaled_X_train.min()
```

## Building the model

```python
from keras.models import Sequential
from keras.layers import Dense
```

### Model Architecture

```python
model = Sequential()

model.add(Dense(4, input_dim=4, activation="relu")) #? input_dim: The number of features

model.add(Dense(8, activation="relu"))

model.add(Dense(1, activation="sigmoid"))
```

### Compiling the model

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

### Training the model

```python
model.fit(scaled_X_train,
          y_train,
          epochs=50,
          verbose=2)
```

## Evaluating the model

```python
predictions = (model.predict(scaled_X_test) > 0.5).astype("int32")
predictions
```

```python
model.metrics_names
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions))
print()
print(classification_report(y_test, predictions))
```

## Saving the Model

```python
model.save("mymodel.h5")
```

## Loading the Model

```python
from tensorflow import keras
new_model = keras.models.load_model("mymodel.h5")
```

```python
predictions = (new_model.predict(scaled_X_test) > 0.5).astype("int32")
predictions
```
